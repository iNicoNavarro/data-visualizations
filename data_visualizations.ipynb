```python
import requests
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
from io import StringIO
```

```python
data_url = "https://raw.githubusercontent.com/mattharrison/datasets/master/data/alta-noaa-1980-2019.csv"
response = requests.get(data_url)
if response.status_code == 200: 
    raw_data = response.content.decode('utf-8')
    data = StringIO(raw_data)
    df_raw = pd.read_csv(data)
    
```

```python
df_raw.info()
```

```python
#Plotting a histogram of snow depth

plt.figure(figsize=[10, 6])
plt.hist(
    df_raw['SNWD'].dropna(),
    bins=30,
    edgecolor='k',
    alpha=0.8    
)
plt.title('Histogram of snow depth')
plt.xlabel('Snow Depth (SNWD)')
plt.ylabel('Frecuency')
plt.show()
```

```python
# Plot a scatter plot of snowfall versus temperature.

plt.figure(figsize=[10, 6])
plt.scatter(
    df_raw['TMAX'],
    df_raw['SNOW'],
    alpha=0.4,   
)
plt.title('Snowfall vs Temperature')
plt.xlabel('Temperature (TMAX)')
plt.ylabel('Snowfalls (SNOW)')
plt.show()
```

```python
# PLot a line for each of the following temperatures: Mininum, Observation and Maximun.

plt.figure(figsize=[15, 6])
plt.plot(
    pd.to_datetime(df_raw['DATE']),
    df_raw['TMIN'],
    label='Minimum Temp (TMIN)', 
    alpha=0.7
)
plt.plot(
    # pd.to_datetime(df_raw['DATE']),
    df_raw['TOBS'],
    label='Observated Temp (TOBS)', 
    alpha=0.7
)
plt.plot(
    pd.to_datetime(df_raw['DATE']),
    df_raw['TMAX'],
    label='Maximum Temp (TMAX)', 
    alpha=0.7
)
plt.title('Temperatures Minimum, Observated y Maximum along of time')
plt.xlabel('Date')
plt.ylabel('Temperature (°F)')
plt.legend()
plt.show()
```
